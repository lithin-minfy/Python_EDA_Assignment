```python
# Load a CSV file and display the first 5 rows.
import numpy as np
import pandas as pd

df = pd.read_csv('/content/oecd_bli_2015.csv')
df.head()
```

```python
# Find missing values in a DataFrame and count them column-wise.

missing_values_count = df.isnull().sum()
print(missing_values_count)
```

```python
# Filter rows in a DataFrame where a column value > 100.
filtered_df = df[df['Value'] > 100]
print(filtered_df)
```

```python
# Group a DataFrame by a column and calculate the mean of each group.
grouped = df.groupby('Country')['Value'].mean()
print(grouped)
```

```python
# Sort a DataFrame by multiple columns.
df.sort_values(by=['Country', 'Value'],ascending = (True,False),inplace=True)
print("sorted columns:")
print(df[['Country','Value']])
```

```python
# Add a new column to a DataFrame that is a transformation of existing columns.
df['total'] = df['Value'] * df['Value']
print(df[['Country','Value','total']])
```
